#### Conflict 2: Category not match
Categories of dataset2 does not match with dataset1

```python
print(df[df['SourceName'] == 'www.jobhuntlisting.com']['Category'].value_counts())
print(df[df['SourceName'] != 'www.jobhuntlisting.com']['Category'].value_counts())
```

create a category map and replace them

```python
# Code to fix the conflict

category_map = {
    'Information Technology':'IT Jobs',
    'Health':'Healthcare & Nursing Jobs',
    'Engineering':'Engineering Jobs',
    'Finance':'Accounting & Finance Jobs',
    'Sales':'Sales Jobs',
    'Hospitality':'Hospitality & Catering Jobs',
    'Education':'Teaching Jobs',
    'Marketing':'PR, Advertising & Marketing Jobs'
}
for cat1 in category_map.keys():
    idx = df[df['Category'] == cat1].index
    df.loc[idx, 'Category'] = category_map[cat1]
```

```python
df[df['SourceName'] == 'www.jobhuntlisting.com']['Category'].value_counts()
```

#### Conflict 3: ContractTime not match
ContractTime of dataset2 does not match with dataset1

```python
print(df[df['SourceName'] == 'www.jobhuntlisting.com']['ContractTime'].value_counts())
print(df[df['SourceName'] != 'www.jobhuntlisting.com']['ContractTime'].value_counts())
```

create a category map and replace them and replace empty value as non-specified

```python
contract_time_map = {
    'Permanent':'permanent',
    'Fixed Term Contract':'contract'
}
contract_time_list = ['permanent', 'contract', 'non-specified']
for cont1 in contract_time_map.keys():
    idx = df[df['ContractTime'] == cont1].index
    df.loc[idx, 'ContractTime'] = contract_time_map[cont1]
df['ContractTime'] = df['ContractTime'].apply(lambda x: x if x in contract_time_list else 'non-specified')
```

```python
print(df[df['SourceName'] == 'www.jobhuntlisting.com']['ContractTime'].value_counts())
print(df[df['SourceName'] != 'www.jobhuntlisting.com']['ContractTime'].value_counts())
```
